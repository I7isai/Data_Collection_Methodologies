**Problem:**

Using only wikipedia, and taking the example of American presidents, build a script or notebook in Python that produces a JSON file with the personal data section of each US president


Before starting it is necessary to import some libraries

```python
import requests
import numpy as np
from bs4 import BeautifulSoup
from bs4.element import Tag
import pandas as pd
import requests,json
```

**1.**

In a first approach we need a dictionary with the personal page of each USA president.

For this, we will reuse the code made during classes which results in a dictionary in each the keys represent the name of each US president and the values, the wikipedia pages of each president.

```python
listpresidents = requests.get('https://en.wikipedia.org/wiki/President_of_the_United_States#Timeline_of_presidents').text
```

```python
'Joe Biden' in listpresidents
```

```python
len(listpresidents)
```

```python
soup_presidents = BeautifulSoup(listpresidents, 'html.parser')
```

```python
print(soup_presidents.prettify()[:2000])
```

```python
for h2 in soup_presidents.find_all('h2'):
  s = h2.find('span')
  if s:
    print(s.text)
```

```python
soup_presidents.find_all("table", attrs={'class': 'wikitable'})
```

```python
d = {}
for table in soup_presidents.find_all("table", attrs={'class': 'wikitable'}):
    rows = table.find_all('tr')
    if len(rows) != 8: continue
    for row in rows:
        for i, cell in enumerate(row.find_all('td')):
            if i != 4: continue
            for a in cell.children:
                if type(a) is Tag and a.name == 'a':
#                  #d[a.text] = f"https://wikipedia.com{a.attrs['href']}"
                  d[a.text.replace(u'\xa0', u' ')] = f"https://wikipedia.com{a.attrs['href']}"
```

```python
d
```

**2.**

With the code used in the classes, we got a "d" dictionary with all the US presidents and the respective wikipedia page.

Then we will iterate the dictionary "d" and for each key, we will go to the personal page to retrieve the personal information of each president.

```python
lista_potus = []
for key,value in zip(d.keys(),d.values()):

    personal_details = pd.read_html(requests.get(d[key]).text)[0].set_axis([key , 'details'], axis=1).set_index(key, drop=True).loc['Personal details': 'Signature'].to_dict()#.update({'Signature': None})
    personal_details['details']['Signature'] = 'none'  
    
    lista_potus.append(personal_details)
```

The code above "pd.read_html(requests.get(d[key]).text)" returns a list of all tables on each page. The Table (infotable) with the personal information of each president is in position [0] so we will always look for that table.

Because some presidents do not have the name, but the military title as column identifier, we changed the column titles to the president's name (key) and details.

Then we changed the index, so that it would be the personal details of each president (Name, Age, etc...).

Finally, as we are only interested in the personal details, we filter the table by index, where we only want the personal details up to the signature. Finally, we converted the collected information into a dictionary and appended it to the list "potus_list".

Since the signature was an image, which resulted in NaN (which conflicted with the JSON format), we changed the NaN to "none".

Below we have the list resulting from the code.

**3.**

Finally, we dumped the potus_list list to a JSON file named "potus_details.json"

```python
with open('potus_details.json', 'w') as fout:
    json.dump(lista_potus , fout)
```

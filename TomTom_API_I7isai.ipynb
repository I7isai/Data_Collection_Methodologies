```python
import requests

url = "https://api.tomtom.com/search/2/geometrySearch/cafe.json?key=key&geometryList=[{\"type\":\"POLYGON\", \"vertices\":[\"38.71288343038636,-9.139416657257703\",\"38.71356048043736,-9.136598996737776\",\"38.708751403434285,-9.134634063679362\",\"38.70804550456099,-9.138156043150783\"]}]&categorySet=9376&limit=100&language=pt-PT"

payload = ""
headers = {
  'Authorization': 'Basic amxjb2l0b0BnbWFpbC5jb206UEdtb2R1bG82IQ=='
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)

#(Note: the key parameter is a personal key provided by TomTom)
```

**After obtaining the results, we will proceed with their analysis.**



```python
import pandas as pd
import json
```

```python
data1 = json.loads(requests.get(url).text)['results']
data1
```

```python
for key in data1:
  print(key['poi']['name'],'-',key['address']['freeformAddress'])
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Polygon, LineString, Point
```

In this phase we will test if all the coffees obtained are located in Baixa Pombalina.
For this purpose, it will be necessary to obtain the latitudes and longitudes of all the coffees that are made afterwards.

Validation will begin with the survey of the coordinates of each cafe. The goal is to get the GeoDataFrame **cafemoradas_gdf** with the two columns of the DataFrame **cafemoradas** and with three more columns *latitude*, *longitude* and *geometry* that displays the coordinate pair

```python
cafes_lat_log = pd.DataFrame(columns = ['placeId', 'nome', 'morada', 'latitude','longitude'])
for key in data1:
    cafes_lat_logRow = {'placeId':key['id'], 'nome':key['poi']['name'], 'morada':key['address']['freeformAddress'], 'latitude':key['position']['lat'],'longitude':key['position']['lon']}
    cafes_lat_log = cafes_lat_log.append(cafes_lat_logRow, ignore_index = True)
cafes_lat_log

```

```python
cafesmoradas_gdf = gpd.GeoDataFrame(cafes_lat_log, geometry = gpd.points_from_xy(cafes_lat_log['longitude'],cafes_lat_log['latitude']))

cafesmoradas_gdf
```

Next, the objective is to obtain a Polygon that represents the Baixa Pombalina area of ​​Lisbon. The coordinates used were presented at the beginning of this report.

```python
baixa_longitude = [-9.139416657257703, -9.136598996737776, -9.134634063679362, -9.138156043150783]
baixa_latitude = [38.71288343038636, 38.71356048043736, 38.708751403434285, 38.70804550456099]

baixa_pombalina_area = Polygon(zip(baixa_longitude, baixa_latitude))
```

Now all the conditions for carrying out the test are met. The validation will be performed by applying the *within* method to the GeoDataFram columne *cafemoradas_gdf['geometry']*.

```python
cafesmoradas_gdf['geometry'].within(baixa_pombalina_area)
```

Through the validation carried out, it is possible to conclude that the coffees introduced in the DataFrame **cafesmoradas** are located in Baixa Pombalina in Lisbon.

Additionally, a search was carried out on google maps, for each of the cafes found in order to complement the verification that they are in downtown Pombalina.
